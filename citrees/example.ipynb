```python
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import confusion_matrix
from citrees import CITreeClassifier
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, BaseDecisionTree
from dtreeviz.trees import dtreeviz
```

```python
from sklearn.tree._tree import Tree
```

```python
data = load_breast_cancer()
X = data.data
y = data.target
```

```python
import numpy as np
CITreeClassifier.tree_ = Tree(X.shape[1], np.array([0,1]), X.shape[0])
```

```python
clf = CITreeClassifier(random_state=2, selector='hybrid')
clf.fit(X,y)
y_pred = clf.predict(X)
display(confusion_matrix(y, y_pred))
pd.DataFrame({
    'features':data.feature_names,
    'importance':clf.feature_importances_
}).sort_values('importance',ascending=False)
```
